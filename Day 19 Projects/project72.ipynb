```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return x**2 + y**2

def gradient(x, y):
    fx = 2 * x
    fy = 2 * y
    return fx, fy
```

```python
x = np.linspace(-3, 3, 20)
y = np.linspace(-3, 3, 20)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
FX, FY = gradient(X, Y)
```

```python
plt.figure(figsize=(8, 6))

# Contour plot (level curves)
contours = plt.contour(X, Y, Z, levels=20, cmap='coolwarm')
plt.clabel(contours, inline=True, fontsize=8)

# Quiver plot for gradients
plt.quiver(X, Y, FX, FY, color='black')

plt.title("Gradient Visualization on Contour Plot of f(x, y) = x + y")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid(True)
plt.axis('equal')
plt.show()
```
