```python
import numpy as np
```

```python
def f(x, y):
    return 2 * x**2 + 3 * x * y + y**3
```

```python
def partial_x(x, y):
    return 4 * x + 3 * y

def partial_y(x, y):
    return 3 * x + 3 * y**2
```

```python
x, y = 1, 2

fx = partial_x(x, y)
fy = partial_y(x, y)
f_value = f(x, y)

print("Function value f(x, y):", f_value)
print("Partial derivative with respect to x (∂f/∂x):", fx)
print("Partial derivative with respect to y (∂f/∂y):", fy)
```
