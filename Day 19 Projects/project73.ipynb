```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

```

```python
# Define the function f(x, y)
def f(x, y):
    return 2 * x**3 + 3 * y
```

```python
x = np.linspace(-5, 5, 100)
```

```python
y = 1
```

```python
z = f(x, y)
```

```python
# Plot the function
plt.figure(figsize=(8, 5))
plt.plot(x, z, color='blue', label='f(x, y=1) = 2x³ + 3y')
plt.title('2D Plot of f(x, y) = 2x³ + 3y')
plt.xlabel('x')
plt.ylabel('f(x, y)')
plt.grid(True)
plt.legend()
plt.show()
```
