```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Define the function
def f(x, y):
    return 4 * x**2 * y + 2 * y**3 * x
```

```python
# Create grid of x and y values
x = np.linspace(-3, 3, 100)
y = np.linspace(-3, 3, 100)
X, Y = np.meshgrid(x, y)
```

```python
# Compute f(x, y)
Z = f(X, Y)
```

```python
# Create a 3D surface plot
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(111, projection='3d')
```

```python
# Plot the surface
surface = ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none', alpha=0.9)
```

```python
# Labels and title
ax.set_title('3D Surface Plot of f(x, y) = 4xy + 2yx', fontsize=13)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
```

```python
# Add color bar for clarity
fig.colorbar(surface, shrink=0.6, aspect=10)

plt.show()
```
