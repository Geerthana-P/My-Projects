```python

import pandas as pd

```

```python
# 1Ô∏è‚É£ Create Employee Details DataFrame
employees = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'department': ['HR', 'Finance', 'IT', 'HR', 'Finance']
})

print("üè¢ Employee Details:")
print(employees, "\n")
```

```python
# 2Ô∏è‚É£ Create Performance Review DataFrame
performance = pd.DataFrame({
    'employee_id': [101, 102, 103, 104, 105],
    'performance_score': [85, 90, 78, 92, 88]
})
```

```python
print("üìä Performance Review Data:")
print(performance, "\n")
```

```python
# 3Ô∏è‚É£ Merge both DataFrames on 'employee_id'
merged_data = pd.merge(employees, performance, on='employee_id', how='inner')

print("‚úÖ Merged Employee + Performance Data:")
print(merged_data, "\n")
```

```python
# 4Ô∏è‚É£ Calculate Average Performance Score per Department
avg_performance = merged_data.groupby('department')['performance_score'].mean().reset_index()

# Rename columns for clarity
avg_performance.columns = ['Department', 'Average_Performance_Score']

print("üìà Average Performance Score per Department:")
print(avg_performance)
```
