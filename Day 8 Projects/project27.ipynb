```python

import pandas as pd
import numpy as np
```

```python
# 1Ô∏è‚É£ Create movie ratings DataFrame (with messy titles and missing ratings)
movie_ratings = pd.DataFrame({
    'Movie_Title': [' Avengers ', 'avatar', 'The Dark Knight ', 'AVATAR ', ' avengers', 'Interstellar'],
    'User_ID': [1, 2, 3, 4, 5, 6],
    'Rating': [5, np.nan, 4.5, 4.0, np.nan, 4.8]
})

print("üé¨ Original Movie Ratings Data:")
print(movie_ratings, "\n")
```

```python
# 2Ô∏è‚É£ Clean movie titles ‚Äî make lowercase and remove extra spaces
movie_ratings['Movie_Title'] = movie_ratings['Movie_Title'].str.lower().str.strip()

print("üé¨ After Standardizing Movie Titles:")
print(movie_ratings, "\n")
```

```python
# Fill missing ratings with the median rating (instead of average this time)
median_rating = movie_ratings['Rating'].median()
movie_ratings['Rating'] = movie_ratings['Rating'].fillna(median_rating)

print("‚≠ê After Filling Missing Ratings with Median:")
print(movie_ratings, "\n")
```

```python
# 4Ô∏è‚É£ Create movie information DataFrame
movie_info = pd.DataFrame({
    'Movie_Title': ['avengers', 'avatar', 'the dark knight', 'interstellar'],
    'Genre': ['Action', 'Sci-Fi', 'Action', 'Sci-Fi'],
    'Release_Year': [2012, 2009, 2008, 2014]
})

print("üé• Movie Info Dataset:")
print(movie_info, "\n")
```

```python
# 5Ô∏è‚É£ Merge the cleaned movie ratings with movie info
final_data = pd.merge(movie_ratings, movie_info, on='Movie_Title', how='left')

print("‚úÖ Final Cleaned & Merged Dataset:")
print(final_data, "\n")
```

```python
# 6Ô∏è‚É£ Simple analysis: Average rating per genre
avg_genre_rating = final_data.groupby('Genre')['Rating'].mean().reset_index()

print("üéûÔ∏è Average Rating per Genre:")
print(avg_genre_rating)
```
