```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Step 1: Create sample data
x = np.linspace(0, 10, 50)
y = np.sin(x)

# Step 2: Create figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Step 3: Line chart on the first subplot
axes[0].plot(x, y, color='blue', marker='o')
axes[0].set_title('Line Chart')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

# Step 4: Scatter plot on the second subplot
axes[1].scatter(x, y, color='red')
axes[1].set_title('Scatter Plot')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')

# Step 5: Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
