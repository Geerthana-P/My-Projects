```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Create DataFrame
data = {
    'Category': ['Electronics', 'Clothing', 'Groceries', 'Furniture', 'Toys'],
    'Sales': [150000, 95000, 120000, 80000, 60000]
}
df = pd.DataFrame(data)

# Step 2: Aggregate Data
category_sales = df.groupby('Category')['Sales'].sum()

# Step 3: Create subplots for bar and pie charts
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar Chart
axes[0].bar(category_sales.index, category_sales.values, color='skyblue')
axes[0].set_title('Total Sales per Product Category')
axes[0].set_xlabel('Category')
axes[0].set_ylabel('Total Sales')
axes[0].tick_params(axis='x', rotation=45)

# Pie Chart
axes[1].pie(category_sales.values, labels=category_sales.index, autopct='%1.1f%%', startangle=90)
axes[1].set_title('Market Share by Category')

# Adjust layout and display
plt.tight_layout()
plt.show()
```
