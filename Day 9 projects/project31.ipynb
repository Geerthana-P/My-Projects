```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the Titanic dataset
titanic = sns.load_dataset('titanic')

# Display first few rows
print(titanic.head())

# Step 2: Survival Bar Plot (Survival rate by passenger class)
plt.figure(figsize=(8, 5))
sns.barplot(x='class', y='survived', data=titanic, palette='Blues')
plt.title('Average Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Step 3: Survival by Age and Gender (Violin Plot)
plt.figure(figsize=(8, 5))
sns.violinplot(x='sex', y='age', hue='survived', data=titanic, split=True, palette='Set2')
plt.title('Age Distribution by Gender and Survival Status')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```
