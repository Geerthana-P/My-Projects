```python
%matplotlib inline
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Step 1: Create a sample dataset
np.random.seed(10)
data = {
    'school': np.random.choice(['School A', 'School B', 'School C'], 150),
    'score': np.random.normal(75, 10, 150)  # average score ~75, std dev ~10
}
df = pd.DataFrame(data)

# Step 2: Histogram of all test scores
plt.figure(figsize=(8, 5))
sns.histplot(df['score'], bins=15, kde=True, color='skyblue')
plt.title('Overall Test Score Distribution')
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.show()

# Step 3: Boxplot comparing scores by school
plt.figure(figsize=(8, 5))
sns.boxplot(x='school', y='score', data=df, palette='Set2')
plt.title('Test Score Comparison by School')
plt.xlabel('School')
plt.ylabel('Score')
plt.show()
```
