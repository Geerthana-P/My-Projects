```python
import numpy as np
```

```python
# Z-score formula: Z = (x - μ) / σ
def calculate_z_score(x, mean, std_dev):
    """Calculate the Z-score for a given value, mean, and standard deviation."""
    z = (x - mean) / std_dev
    return z
```

```python
# Generate random data from a normal distribution
mean = 170
std_dev = 5
data = np.random.normal(mean, std_dev, 1000)

# Display sample statistics
print(f"Sample Mean: {np.mean(data):.2f}")
print(f"Sample Std Dev: {np.std(data):.2f}")
```

```python
# Choose a specific value from the dataset (e.g., 180 cm height)
x_value = 180

# Calculate the Z-score
z_score = calculate_z_score(x_value, np.mean(data), np.std(data))

print(f"Value (x): {x_value}")
print(f"Z-score: {z_score:.2f}")
```

```python
# Interpretation of Z-score
if z_score > 0:
    print(f"The value {x_value} is {abs(z_score):.2f} standard deviations ABOVE the mean.")
elif z_score < 0:
    print(f"The value {x_value} is {abs(z_score):.2f} standard deviations BELOW the mean.")
else:
    print(f"The value {x_value} is exactly at the mean.")
```
