```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define parameters for the normal distribution
mean = 170       # μ (average height, for example)
std_dev = 5      # σ (standard deviation)
n = 1000         # number of data points

# Generate random data following a normal distribution
data = np.random.normal(mean, std_dev, n)

# Display first few data points
data[:10]
```

```python
# Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data)

print(f"Specified Mean (μ): {mean}")
print(f"Specified Std Dev (σ): {std_dev}")
print(f"Calculated Mean: {sample_mean:.2f}")
print(f"Calculated Std Dev: {sample_std:.2f}")
```

```python
# 1σ, 2σ, 3σ ranges from the mean
within_1_std = np.sum((data >= sample_mean - sample_std) & (data <= sample_mean + sample_std))
within_2_std = np.sum((data >= sample_mean - 2*sample_std) & (data <= sample_mean + 2*sample_std))
within_3_std = np.sum((data >= sample_mean - 3*sample_std) & (data <= sample_mean + 3*sample_std))

# Calculate percentages
pct_1_std = (within_1_std / n) * 100
pct_2_std = (within_2_std / n) * 100
pct_3_std = (within_3_std / n) * 100

print(f"Within 1 Std Dev: {pct_1_std:.2f}% (Expected ≈ 68%)")
print(f"Within 2 Std Dev: {pct_2_std:.2f}% (Expected ≈ 95%)")
print(f"Within 3 Std Dev: {pct_3_std:.2f}% (Expected ≈ 99.7%)")
```

```python
plt.figure(figsize=(10,6))
sns.histplot(data, bins=30, kde=True, color='skyblue', edgecolor='black')

plt.title('Normal Distribution of Generated Data', fontsize=14)
plt.xlabel('Value', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
