```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the Poisson parameter (average rate of arrivals)
lambda_val = 5  # average number of customers per minute

# Number of time intervals to simulate
num_intervals = 1000

# Simulate Poisson-distributed data
arrivals = np.random.poisson(lambda_val, num_intervals)

# Display first few simulated values
arrivals[:20]
```

```python
# Create a histogram using Seaborn
plt.figure(figsize=(10,6))
sns.histplot(arrivals, bins=range(0, 15), kde=False, color='skyblue', edgecolor='black')

plt.title('Simulation of Customer Arrivals (Poisson Distribution)', fontsize=14)
plt.xlabel('Number of Customers Arriving per Minute', fontsize=12)
plt.ylabel('Frequency (Number of Intervals)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# Calculate and display mean and variance
sample_mean = np.mean(arrivals)
sample_var = np.var(arrivals)

print(f"Expected (): {lambda_val}")
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Variance: {sample_var:.2f}")

# In a Poisson distribution, mean  variance  
```
