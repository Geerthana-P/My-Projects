```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)

# Two features (x1, x2)
X1 = np.random.rand(100, 1) * 10   # random values 0â€“10
X2 = np.random.rand(100, 1) * 10

# Combine into one matrix (100 x 2)
X = np.hstack((X1, X2))

# True relationship + noise
noise = np.random.randn(100, 1) * 2
y = 2 * X1 + 3 * X2 + 5 + noise
```

```python
# Add bias term (column of ones)
X_b = np.hstack((X, np.ones((X.shape[0], 1))))  # shape (100, 3)

# Initialize parameters randomly
theta = np.random.randn(3, 1)

# Hyperparameters
learning_rate = 0.001
epochs = 1000
```

```python
n = len(X_b)
loss_history = []

for epoch in range(epochs):
    y_pred = X_b @ theta  # Predictions
    
    # Compute gradient
    gradient = (2/n) * X_b.T @ (y_pred - y)
    
    # Update parameters
    theta = theta - learning_rate * gradient
    
    # Compute loss (Mean Squared Error)
    loss = np.mean((y - y_pred)**2)
    loss_history.append(loss)

print(f"Final parameters: m1 = {theta[0][0]:.3f}, m2 = {theta[1][0]:.3f}, b = {theta[2][0]:.3f}")
```

```python
plt.plot(range(epochs), loss_history, color='purple')
plt.title("Loss Convergence over Epochs")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.grid(True)
plt.show()
```
