```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**2

def df(x):
    return 2*x
```

```python
def gradient_descent(x_start, learning_rate, epochs):
    x = x_start
    x_path = [x]
    
    for i in range(epochs):
        grad = df(x)
        x = x - learning_rate * grad
        x_path.append(x)
        
    return np.array(x_path)
```

```python
x_start = 8   # start far from the minimum
epochs = 20

learning_rates = [0.01, 0.1, 1.0]
paths = {}

for lr in learning_rates:
    paths[lr] = gradient_descent(x_start, lr, epochs)
```

```python
x_vals = np.linspace(-10, 10, 200)
y_vals = f(x_vals)

plt.figure(figsize=(8,6))
plt.plot(x_vals, y_vals, 'b', label='f(x) = x')

# Plot each learning rate path
colors = ['red', 'green', 'orange']
for lr, color in zip(learning_rates, colors):
    x_path = paths[lr]
    plt.plot(x_path, f(x_path), marker='o', color=color, label=f'lr={lr}')

plt.title("Gradient Descent Convergence with Different Learning Rates")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
