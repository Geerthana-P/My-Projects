```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random linear data
np.random.seed(42)  # for reproducibility

X = np.linspace(0, 10, 50)
noise = np.random.randn(50) * 2
y = 2 * X + 5 + noise
```

```python
m = np.random.randn()   # random slope
b = np.random.randn()   # random intercept
learning_rate = 0.01
epochs = 1000

# For tracking
loss_history = []
m_history = []
b_history = []
```

```python
n = len(X)

for epoch in range(epochs):
    y_pred = m * X + b
    
    # Compute gradients
    dm = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)
    
    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db
    
    # Compute loss (MSE)
    loss = np.mean((y - y_pred)**2)
    
    # Store history
    loss_history.append(loss)
    m_history.append(m)
    b_history.append(b)

print(f"Final values → m = {m:.3f}, b = {b:.3f}")
```

```python
# Plot data and regression line
plt.figure(figsize=(12,5))

# Left plot: regression fit
plt.subplot(1, 2, 1)
plt.scatter(X, y, color='blue', label='Data')
plt.plot(X, m*X + b, color='red', label='Fitted Line')
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)

# Right plot: loss curve
plt.subplot(1, 2, 2)
plt.plot(range(epochs), loss_history, color='purple')
plt.title("Loss Over Epochs")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.grid(True)

plt.tight_layout()
plt.show()
```
