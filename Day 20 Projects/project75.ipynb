```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Function and its derivative
def f(x):
    return x**2

def df(x):
    return 2*x
```

```python
# Initialization
x = np.random.uniform(-10, 10)   # Random start between -10 and 10
learning_rate = 0.1
epochs = 25

x_history = [x]  # To store x values for visualization
```

```python
for i in range(epochs):
    grad = df(x)
    x = x - learning_rate * grad
    x_history.append(x)

print("Minimum found near x =", x)
```

```python
# Plot the function and the descent path
x_vals = np.linspace(-10, 10, 200)
y_vals = f(x_vals)

plt.plot(x_vals, y_vals, label='f(x) = xÂ²')
plt.scatter(x_history, f(np.array(x_history)), color='red', label='Steps')
plt.plot(x_history, f(np.array(x_history)), color='red', linestyle='--')

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('1D Gradient Descent on a Parabola')
plt.legend()
plt.grid(True)
plt.show()
```
