```python
import numpy as np
```

```python
# Coefficient matrix A
A = np.array([[2, 3],
              [5, -1]])

# Constants vector b
b = np.array([8, 2])
```

```python
# Solve Ax = b
solution = np.linalg.solve(A, b)

x, y = solution
print(f"Solution:\nx = {x}\ny = {y}")
```

```python
# Verify by plugging back into original equations
print("Check first equation:", 2*x + 3*y)  # Should be 8
print("Check second equation:", 5*x - y)   # Should be 2
```
