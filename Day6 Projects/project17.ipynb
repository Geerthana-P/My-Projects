```python
# Linear Regression from Scratch using Normal Equation
import numpy as np
import matplotlib.pyplot as plt

# Optional: display plots inline
%matplotlib inline
```

```python
# 1. Generate synthetic data
np.random.seed(0)  # For reproducibility

X = np.random.rand(100, 1)          # 100 samples, 1 feature
y = 2 * X.flatten() + 3 + np.random.randn(100) * 0.5  # y = 2*X + 3 + noise

# Display first 5 samples
print("First 5 X values:\n", X[:5])
print("First 5 y values:\n", y[:5])
```

```python
# Add bias term (column of ones) to X
X_b = np.c_[np.ones((100, 1)), X]  # shape: (100, 2)

# Display first 5 rows with bias term
print("First 5 X with bias term:\n", X_b[:5])
```

```python
# 2. Apply Normal Equation to find coefficients
beta = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
intercept, slope = beta

print(f"Intercept: {intercept:.3f}")
print(f"Slope: {slope:.3f}")
```

```python
# 3. Make predictions
predictions = X_b @ beta

# Display first 5 predicted values
print("First 5 predicted y values:\n", predictions[:5])
```

```python
# 4. Visualize data and regression line
plt.figure(figsize=(8,5))
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, predictions, color='red', linewidth=2, label='Regression line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression using Normal Equation')
plt.legend()
plt.show()
```
