```python
import numpy as np
```

```python
# Example dataset: 5 samples, 3 features
X = np.array([[1, 200, 3],
              [2, 210, 6],
              [3, 190, 9],
              [4, 220, 12],
              [5, 205, 15]])

print("Original Dataset:\n", X)
```

```python
# Compute mean and standard deviation for each column
mean_X = X.mean(axis=0)
std_X = X.std(axis=0)

# Standardize dataset
X_standardized = (X - mean_X) / std_X

print("\nStandardized Dataset:\n", X_standardized)
```

```python
print("\nColumn-wise Mean after Standardization:\n", X_standardized.mean(axis=0))
print("Column-wise Std after Standardization:\n", X_standardized.std(axis=0))
```
