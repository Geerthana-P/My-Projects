```python
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Generate synthetic data (e.g., 300 customers, 3 clusters)
X, y_true = make_blobs(n_samples=300, centers=3, cluster_std=1.0, random_state=42)

# Visualize the generated data
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title("Simulated Customer Data")
plt.xlabel("Annual Income (normalized)")
plt.ylabel("Spending Score (normalized)")
plt.show()
```

```python
from sklearn.cluster import KMeans

# Create a KMeans object with 3 clusters
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model to the data
kmeans.fit(X)
```

```python
# Get labels assigned to each point
labels = kmeans.labels_

print("Cluster labels for first 10 points:", labels[:10])
```

```python
# Get cluster centers
centers = kmeans.cluster_centers_

# Plot the clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X', label='Centers')
plt.title("Customer Segments Found by K-Means")
plt.xlabel("Annual Income (normalized)")
plt.ylabel("Spending Score (normalized)")
plt.legend()
plt.show()
```
