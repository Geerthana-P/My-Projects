```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a 100x100 "grayscale image" with random pixel values between 0 and 1
fake_image = np.random.rand(100, 100)

# Display the original fake image
plt.imshow(fake_image, cmap='gray')
plt.title("Original Fake Grayscale Image")
plt.axis('off')
plt.show()
```

```python
# Invert the pixel values: bright â†” dark
inverted_image = 1.0 - fake_image

# Crop the central 50x50 portion of the image
cropped_image = fake_image[25:75, 25:75]
```

```python
# Show the original, inverted, and cropped images side by side
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(fake_image, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(inverted_image, cmap='gray')
plt.title("Inverted")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cropped_image, cmap='gray')
plt.title("Cropped")
plt.axis('off')

plt.show()
```
