```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image_path = 'Flower.png'  # Replace with your image file
img = Image.open(image_path).convert('L')  # Convert to grayscale
image = np.array(img, dtype=np.float32)    # Convert to NumPy array
```

```python
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')  # Hide axis numbers
plt.show()
```

```python
def blur_image(img):
    blurred = np.zeros_like(img)  # Output image with same shape
    rows, cols = img.shape
    
    for i in range(rows):
        for j in range(cols):
            neighbors = []  # Store neighbor pixel values
            
            # Loop over 3x3 neighborhood
            for x in range(max(0, i-1), min(rows, i+2)):
                for y in range(max(0, j-1), min(cols, j+2)):
                    neighbors.append(img[x, y])
            
            # Replace pixel with average of neighbors
            blurred[i, j] = sum(neighbors) / len(neighbors)
    
    return blurred
```

```python
blurred_image = blur_image(image)
```

```python
plt.imshow(blurred_image, cmap='gray')
plt.title("Blurred Image")
plt.axis('off')
plt.show()
```
