```python
import numpy as np

# Create a 2D array (5 rows Ã— 4 columns) with random integers between 10 and 100
arr = np.random.randint(10, 100, size=(5, 4))

print("Original Array:\n", arr)
```

```python
# Calculate the mean for each column
col_mean = arr.mean(axis=0)

print("Mean of each column:", col_mean)
```

```python
# Find the maximum value in each row
row_max = arr.max(axis=1)

print("Maximum value in each row:", row_max)
```

```python
# Calculate the total sum of all elements in the array
total_sum = arr.sum()

print("Total sum of all elements:", total_sum)
```

```python
# Select the first two rows and last two columns
subset = arr[0:2, 2:4]
print("First two rows and last two columns:\n", subset)

# Use boolean indexing to find elements greater than 50
greater_than_50 = arr[arr > 50]
print("\nElements greater than 50:", greater_than_50)
```

```python
print("Original Array:\n", arr)
print("\nColumn Means:", col_mean)
print("Row Maximums:", row_max)
print("Total Sum:", total_sum)
print("\nSubset (First 2 rows, Last 2 columns):\n", subset)
print("\nElements greater than 50:", greater_than_50)
```
