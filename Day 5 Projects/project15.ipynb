```python
import numpy as np
```

```python
# Create a 2D array (5 samples × 4 features) with random integers between 10 and 100
arr = np.random.randint(10, 100, size=(5, 4))
print("Original Array:\n", arr)
```

```python
# Step 1: Calculate mean of each column
mean = arr.mean(axis=0)
print("\nMean of each column:\n", mean)

# Step 2: Calculate standard deviation of each column
std = arr.std(axis=0)
print("\nStandard deviation of each column:\n", std)

# Step 3: Normalize (subtract mean, divide by std)
normalized_arr = (arr - mean) / std
print("\nNormalized Array:\n", normalized_arr)
```

```python
# After normalization, each column should have mean ≈ 0 and std ≈ 1
print("Mean after normalization:", normalized_arr.mean(axis=0))
print("Std after normalization:", normalized_arr.std(axis=0))
```
