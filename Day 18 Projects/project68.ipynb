```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def f(x, y):
    return x**2 + y**2

def gradient(x, y):
    return np.array([2*x, 2*y])
```

```python
x = np.linspace(-3, 3, 50)
y = np.linspace(-3, 3, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
x0, y0 = 1, 1  # chosen point
grad = gradient(x0, y0)
print(f"Gradient at ({x0}, {y0}) = {grad}")
```

```python
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot surface
ax.plot_surface(X, Y, Z, cmap='viridis', alpha=0.7)

# Plot the chosen point
ax.scatter(x0, y0, f(x0, y0), color='red', s=50, label="Chosen Point")

# Plot gradient vector (arrow)
ax.quiver(
    x0, y0, f(x0, y0),       # starting point
    grad[0], grad[1], 0,     # direction (dz=0, stays in tangent plane)
    color='black', length=1, arrow_length_ratio=0.2, label="Gradient Vector"
)

ax.set_title("3D Visualization of Gradient of f(x, y) = x + y")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("f(x, y)")
ax.legend()
plt.show()
```
