```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create synthetic data
np.random.seed(42)
X = np.linspace(0, 10, 50)             # 50 points between 0 and 10
y = 3 * X + 4 + np.random.randn(50) * 2  # true line + noise
```

```python
def mse(y, y_pred):
    return np.mean((y - y_pred) ** 2)
```

```python
m = np.random.randn()  # random slope
b = np.random.randn()  # random intercept
learning_rate = 0.01
epochs = 1000
```

```python
n = len(X)

for epoch in range(epochs):
    y_pred = m * X + b

    # Compute gradients
    dm = (-2/n) * np.sum(X * (y - y_pred))
    db = (-2/n) * np.sum(y - y_pred)

    # Update parameters
    m = m - learning_rate * dm
    b = b - learning_rate * db

    # Print loss every 100 epochs
    if epoch % 100 == 0:
        loss = mse(y, y_pred)
        print(f"Epoch {epoch}: m = {m:.3f}, b = {b:.3f}, Loss = {loss:.3f}")
```

```python
# Final predictions
y_pred_final = m * X + b

# Plot
plt.scatter(X, y, color='blue', label='Data points')
plt.plot(X, y_pred_final, color='red', label=f'Best Fit Line: y={m:.2f}x+{b:.2f}')
plt.title("Linear Regression using Gradient Descent")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
