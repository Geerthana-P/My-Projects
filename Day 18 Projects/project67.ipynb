```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x**4 - 3*x**3 + 2

def f_prime(x):
    return 4*x**3 - 9*x**2
```

```python
x = np.random.randn()     # random starting point
learning_rate = 0.01
epochs = 1000

print(f"Initial x: {x:.4f}")
```

```python
x_values = [x]
for i in range(epochs):
    grad = f_prime(x)
    x = x - learning_rate * grad
    x_values.append(x)

    # stop if derivative is nearly zero
    if abs(grad) < 1e-6:
        print(f"Converged at epoch {i}")
        break
```

```python
x_plot = np.linspace(-1, 4, 200)
y_plot = f(x_plot)

plt.figure(figsize=(8,6))
plt.plot(x_plot, y_plot, label="f(x) = x⁴ - 3x³ + 2", color='blue')
plt.scatter(x_values, f(np.array(x_values)), color='red', label="Descent Path")
plt.title("Gradient Descent to Find Minimum of f(x)")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

print(f"Final x ≈ {x:.4f}, f(x) ≈ {f(x):.4f}")
```
