```python
import pandas as pd
```

```python
# Step 2: Create sample purchase data
data = {
    'CustomerID': [101, 102, 103, 101, 102, 104, 103, 101],
    'Product': ['Shoes', 'Bag', 'Watch', 'Shoes', 'Shoes', 'Watch', 'Bag', 'Watch'],
    'Amount': [1200, 800, 1500, 1000, 1200, 2000, 800, 1500]
}

df = pd.DataFrame(data)

print("üßæ Original Customer Purchase Data:")
print(df)
```

```python
# Step 3: Find total spending for each customer
total_spending = df.groupby('CustomerID')['Amount'].sum()

print("üí∞ Total Spending for Each Customer:")
print(total_spending)
```

```python
# Step 4: Find the most popular product
product_popularity = df['Product'].value_counts()

print("üèÜ Product Popularity (Most to Least):")
print(product_popularity)
```

```python
most_popular = product_popularity.idxmax()
print(f"üåü Most Popular Product: {most_popular}")
```
