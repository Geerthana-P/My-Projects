```python
# ğŸ“ Objective: Create a DataFrame for student grades and calculate the average
# score for each student and each subject.

# Step 1: Import pandas
import pandas as pd

# Step 2: Create sample DataFrame
data = {
    'Student': ['Alice', 'Bob', 'Alice', 'Bob', 'Charlie', 'Charlie'],
    'Math': [85, 90, 78, 88, 92, 95],
    'Science': [80, 85, 82, 90, 88, 92],
    'English': [75, 82, 79, 85, 91, 89]
}

df = pd.DataFrame(data)

print("ğŸ“˜ Original Student Grades Data:")
print(df)
print("-" * 60)

# Step 3: Calculate average score for each student
student_avg = df.groupby('Student').mean(numeric_only=True)

print("ğŸ‘©â€ğŸ“ Average Score for Each Student:")
print(student_avg)
print("-" * 60)

# Step 4: Calculate average score for each subject
subject_avg = df.mean(numeric_only=True)

print("ğŸ“š Average Score for Each Subject:")
print(subject_avg)
```
