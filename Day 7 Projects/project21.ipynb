```python
import pandas as pd
```

```python
data = {
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Quantity': [10, 5, 12, 8, 7],
    'Price': [20, 15, 20, 25, 15]
}
sales_df = pd.DataFrame(data)
```

```python
print("ðŸ“˜ Original Sales Data:")
print(sales_df)
```

```python
sales_df['Total Sales'] = sales_df['Quantity'] * sales_df['Price']
```

```python
print("ðŸ’¡ Data with 'Total Sales' column:")
print(sales_df)
```

```python
product_sales = sales_df.groupby('Product')['Total Sales'].sum()
```

```python
print("ðŸ“Š Total Sales per Product:")
print(product_sales)
```

```python
import pandas as pd

# Step 1: Load the CSV
sales_df = pd.read_csv('data.csv')

# Step 2: Clean up column names (important!)
sales_df.columns = sales_df.columns.str.strip()

print("ðŸ“˜ Columns after cleaning:")
print(sales_df.columns)
print("-" * 50)

# Step 3: Add 'Total Sales'
sales_df['Total Sales'] = sales_df['Quantity'] * sales_df['Price']

print("ðŸ’¡ Data with 'Total Sales':")
print(sales_df)
print("-" * 50)

# Step 4: Group by Product
product_sales = sales_df.groupby('Product')['Total Sales'].sum()

print("ðŸ“Š Total Sales per Product:")
print(product_sales)
```
